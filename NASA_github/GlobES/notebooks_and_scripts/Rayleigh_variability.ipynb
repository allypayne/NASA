# Rayleigh Scattering Variability

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
e0_1d_no_rayleigh= np.genfromtxt('./true_psg_files/psg_rad_e0_1d_norayleigh.txt') # 216 calculations, binning=5
e0_1d_rayleigh= np.genfromtxt('./true_psg_files/psg_rad_1d_e0.txt') # 1 calculation
e0_3d_norayleigh= np.genfromtxt('./true_psg_files/psg_rad_e0_3d_norayleigh.txt')
e0_3d= np.genfromtxt('./true_psg_files/psg_rad_3d_e0.txt')
```

```python
fig, ax = plt.subplots(1, 2, sharex='col', sharey='row', figsize=(14,6), dpi=220, facecolor='bisque', alpha=0.5)

# Left Plot
ax[0].plot(e0_1d_no_rayleigh[:,0], e0_1d_no_rayleigh[:,1], color='blue', label='1-D Model: No Rayleigh Scattering')
ax[0].plot(e0_3d_norayleigh[:,0], e0_3d_norayleigh[:,1], color='limegreen', label='3-D Model: No Rayleigh Scattering')
ax[0].grid()
ax[0].set_xlabel('Wavelength [$\mu$m]')
ax[0].set_ylabel('Geometric Albedo')
ax[0].legend()
ax[1].set_title('Rayleigh Scattering')
#Right Plot
ax[1].plot(e0_1d_rayleigh[:,0], e0_1d_rayleigh[:,1], color='blue', label='1-D Model: Rayleigh Scattering Included')
ax[1].plot(e0_3d[:,0], e0_3d[:,1], color='limegreen', label='3-D Model: Rayleigh Scattering Included')
ax[1].grid()
ax[1].set_xlabel('Wavelength [$\mu$m]')
ax[1].set_ylabel('Geometric Albedo')
ax[1].legend()
ax[0].set_title('No Scattering')
#plt.suptitle('3-D VS 1-D Modern Earth Spectra Comparison', size=15);
plt.savefig('Rayleigh_variations.png')
```
