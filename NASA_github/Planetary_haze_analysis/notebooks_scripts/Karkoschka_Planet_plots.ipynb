```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read in Karkoschka data
day_night_jup= np.genfromtxt('./../../jupiter_spectra/coulterbarnesfortney_fig5_jup_dbf5.txt')
karkoschka_fromgv= np.genfromtxt('./../../jupiter_spectra/KarkoschkaAlbedoJupiter.txt')
psg_default_saturn= np.genfromtxt('./../../psg_rad_saturn_plain.txt')
psg_saturn_haze= np.genfromtxt('./../../psg_rad_saturn_haze_ally.txt')
psg_default_saturn_n1_l80= np.genfromtxt('./../../psg_rad_saturn_n1_L80.txt')
# The data for each of the spectral data types (day/night, cloudy/clear) come from day_night_jup. need to sort:
dayside_cloudy= np.array(day_night_jup[0:330])
psg_default_data_jupiter= np.genfromtxt('/Users/apayne3/Desktop/Project_Part2_Colors/jupiter_spectra/psg_rad_wide_default.txt')
random1= np.genfromtxt('psg_rad_random_plot_1.txt')
random2= np.genfromtxt('psg_rad_random2.txt')
psg_uranus_default= np.genfromtxt('./../../psg_rad_uranus_plain.txt')
psg_uranus_rec= np.genfromtxt('./../../psg_rad_uranus_n1_l80.txt')
psg_neptune_plain=  np.genfromtxt('./../../psg_rad_neptune_plain.txt')
psg_neptune_rec=  np.genfromtxt('./../../psg_rad_neptune_n1_l80.txt')
```

```python
plt.figure(figsize=(8,4), dpi=180)
plt.plot((karkoschka_fromgv[:,0])/1000, karkoschka_fromgv[:,3], label='Karkoschka', linewidth=1, color='royalblue')
plt.plot(dayside_cloudy[:,1],dayside_cloudy[:,3], label='Dayside Cloudy CBF (MAIN REFERENCE)', linewidth=0.7, color='dodgerblue')
#plt.plot(random2[:,0],random2[:,1], label='Random 2', linewidth=0.7, color='lime')
plt.plot(psg_default_data_jupiter[:,0], psg_default_data_jupiter[:,1], label='PSG Default', linewidth=0.7, linestyle='--', color='violet')
#plt.plot(random1[:,0], random1[:,1], label='PSG R1', linewidth=0.7, color='red')
plt.grid()
plt.axvline(x=.678, label='Peak I/F occurs at 0.678 microns', color='sandybrown', linestyle='-.', linewidth=0.5)
plt.ylim(0,0.6)
plt.xlim(0,4.8)
plt.title('Zoomed on Karkoschka Data: Jupiter')
plt.xlabel(r'Wavelength [$\mu$m]')
plt.ylabel('I/F')
plt.xticks(np.arange(0,4.8, 0.2))
plt.grid(True, linewidth=0.4, linestyle='--') 
plt.legend(fontsize=7)
plt.xlim(0.3,1)
plt.show()
```
